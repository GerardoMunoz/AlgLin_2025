# Visualización de una Onda en un Sintetizador de Sonido
Para un sintetizador de sonidos se han almacenados diferentes efectos musicales en archivos. En particular se desea visualizar uno de ellos. En el archivo están los datos de la función $f(\alpha)=256*cos(2*pi*\alpha/256)$, donde $\alpha$ es el número del renglón que va dese 0 hasta 255.
Se quiere grficar la curva en una pantalla que tiene 480 de alto (el cero es arriba) y 640 de ancho (el cero es a la izquierda). Las coordenadas de la pantalla las representaremos con $(x,y)$

1. Encuentre la función $g$ para el eje horizontal $x=g(\alpha)$
2. Encuentre la función $h$ para el eje vertical $y=h(f(\alpha))$

image.png



Solución:

Cómo se quieren preservar las proporciones en ambos ejes usaremos la ecuación de la recta $$x=m_1\alpha+b_1$$  $$y=m_2f(\alpha)+b_2$$.

Si conseguimos 2 puntos en cada uno de los problemas es posible encontrar cada recta.


1. Función $x=g(\alpha)$

Comenzaremos hallando dos puntos de la funcion $x=m_1\alpha+b_1$. Sabemos que $\alpha$ va dese 0 hasta 255, además que $x$ va entre 0 y 639. Por lo tanto los dos puntos son:

|    |  $\alpha$ |  $x$   |
|---|---|---|
| Punto 1 | 0   |   0   |
| Punto 2 | 255 | 639 |

Cuando los ceros coinciden, el problema se simplifica a un problema de proporciones o de regla de tres. $$x=\alpha * 639 /255$$ Lo cual implica que en la primera recta $m_1=639/255$ y $b_1=0$


2. Función y=h(f(\alpha))
Continuemos hallando los puntos para la funcion $y=m_2f(\alpha)+b_2$. Sabemos que $f(\alpha)$ toma los valores de la función $256*cos(2*pi*\alpha/256)$, cuando $cos$ es $-1$ entonces $f(\alpha)=-255$ y cuando $cos$ es $1$ entonces $f(\alpha)=255$  por lo tanto $f(\alpha)$ va entre -255 hasta 255. Por otro lado, $y$ va desde 479 (superior) hasta cero (inferior). Por lo tanto los dos puntos son:

|    |  $f(\alpha)$ |  $y$   |
|---|---|---|
| Punto 1 | $f(\alpha_0)=-255$   |   $y_0=0$   |
| Punto 2 | $f(\alpha_1)=255$ | $y_1=479$ |

En este caso, como los ceros no coinciden, es necesario considerar la ecuación de la recta. Utilizaremos la ecaución de la pendiente

$$\frac{y_1 - y_0}{f(\alpha_1)- f(\alpha_0)}=\frac{y - y_0}{f(\alpha)- f(\alpha_0)}$$
$$\frac{479 - 0}{255 - (-255)}=\frac{y - 0}{f(\alpha)- (-255)}$$
$$479(f(\alpha) + 255)=510y$$
$$y=\frac{479}{510}f(\alpha) + \frac{479}{2}$$

Por lo tanto en la segunda recta  $m_2=\frac{479}{510}$ y $b_2=\frac{479}{2}$



```python
# prompt: draw in matplotlib f(\alpha)=256*cos(pi*\alpha/256) for alpha between 0 to 255

import matplotlib.pyplot as plt
import numpy as np

# Define the function f(alpha)
def f(alpha):
  return 256 * np.cos(2*np.pi * alpha / 256)

# Create an array of alpha values from 0 to 255
alpha_values = np.arange(0, 256)

# Calculate the corresponding f(alpha) values
f_alpha_values = f(alpha_values)

# Calculate the x and y coordinates for plotting
x_coordinates = alpha_values #* 639 / 255
y_coordinates = f_alpha_values#(479 / 510) * f_alpha_values + 479 / 2

# Plot the function
plt.plot(x_coordinates, y_coordinates)
plt.xlabel("alpha")
plt.ylabel("f(alpha)")
plt.title("f(alpha) = 256*cos(2*pi*alpha/256)")
plt.show()
```

```python
# prompt: draw in matplotlib f(\alpha)=256*cos(pi*\alpha/256) for alpha between 0 to 255

import matplotlib.pyplot as plt
import numpy as np

# Define the function f(alpha)
def f(alpha):
  return 256 * np.cos(2*np.pi * alpha / 256)

# Create an array of alpha values from 0 to 255
alpha_values = np.arange(0, 256)

# Calculate the corresponding f(alpha) values
f_alpha_values = f(alpha_values)

# Calculate the x and y coordinates for plotting
x_coordinates = alpha_values * 639 / 255
y_coordinates = (479 / 510) * f_alpha_values + 479 / 2

# Plot the function
plt.plot(x_coordinates, y_coordinates)
plt.xlabel("x")
plt.ylabel("y")
plt.title("f(alpha) = 256*cos(2*pi*alpha/256)")
plt.show()
```
