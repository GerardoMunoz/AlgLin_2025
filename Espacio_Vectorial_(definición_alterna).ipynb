# Espacio Vectorial
## Deficnición
Un **Espacio  Vectorial**
Un conjunto $V$ junto con dos operaciones:

* Suma de vectores: $+:V \times V \to V$

* Multiplicación por un escalar: $\cdot: F \times V \to V$

Donde $F$ es un cuerpo, en particular  si $F=\mathbb{R}$ debe cumplir los siguientes axiomas:

## Axiomas  (para cualesquiera $u,v,w \in V$):

1. **Vector Cero**: Existe un vector $\vec{0}$ tal que $\vec{0}=0u$ parta todo $u \in V$

2. **Identidad**: $1u = u$, donde $1$ es el elemento neutro multiplicativo del cuerpo.

3. **Opuesto**: Para cada $u$ existe $-u$ tal que $-u=(-1)u$

4. **Asociatividad** de suma y producto:

* $(u + v) + w = u + (v + w)$

* $(ab)u = a(bu)$ (Esta tiene su matiz ya que a la derecha se asocia un escalar con un vector)

5. **Conmutatividad**: $u + v = v + u$

6 **Distributivas** a izquierda y derecha:

* $a(u+v) = au + av$

* $(a+b)u = au + bu$

## Ejercicio
Demeuestre que estos axiomas implican:
* Existe un vector $\vec{0}$ tal que $u+\vec{0}=u$ (Elemento neutro)
* Para cada $u$ existe $−u$ tal que $u+(−u)=\vec{0}$ (Elemento inverso)
