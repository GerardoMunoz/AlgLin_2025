Punto 1

¿Qué ángulo debe girar en Q para apuntar al punto R? (escribir el coseno del ángulo)


````python
QR=R-Q
QR
````

````python
PQ=Q-P
PQ
````

````python
cos_alpha=(QR.dot(PQ))/(QR.norm()*PQ.norm())
cos_alpha
````
