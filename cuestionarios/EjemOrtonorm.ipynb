Encontrar una base ortogonal con las columnas de B

```python
import sympy as sp
B=sp.Matrix([
  [2, 0, 1],
  [-1, 3 ,2],
  [1,-1, 0]])
B
```

```python
v0=B[:,0]
v0
```

```python
u0=v0
u0
```

```python
v1=B[:,1]
v1
```

```python
u1=v1-u0*(u0.dot(v1))/(u0.dot(u0))
u1
```

```python
u0.dot(u1)
```

```python
v2=B[:,2]
v2
```

```python
u2=v2-u0*(u0.dot(v2))/(u0.dot(u0))-u1*(u1.dot(v2))/(u1.dot(u1))
u2
```

```python
B_og=u0.row_join(u1).row_join(u2)
B_og
```

```python
w0=u0/u0.norm()
w0
```

```python
w1=u1/u1.norm()
w1
```

```python
w2=u2/u2.norm()
w2
```

```python
B_on=w0.row_join(w1).row_join(w2)
B_on
```

```python
B_on@B_on.T
```
