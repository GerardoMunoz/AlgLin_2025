```python
import sympy as sp
```

# $A_0$

```python
A0=sp.Matrix([[1,4],[2,8],[-1,-4]])
M=A0[:,:]
M#sp.latex(M)
```

```python
M[1,:]=M[1,:]-2*M[0,:]
M[2,:]=M[2,:]+M[0,:]
M#sp.latex(M)
```

Para encontrar las columnas que se pueden eliminar buscamos la forma escalón

$A_0=\left[\begin{matrix}1 & 4\\2 & 8\\-1 & -4\end{matrix}\right] \sim \left[\begin{matrix}1 & 4\\0 & 0\\0 & 0\end{matrix}\right]$

Cómo hay una columna con l-pivote (lugar del pivote) entonces:
* el rango ($\rho$) es 1
* sólo hay un vector en la base del espacio columna Co($A_0$), $\left[\begin{matrix}1\\2\\-1\end{matrix}\right]$
* geométricamente el espacio columna Co($A_0$) es una recta en $\mathbb{R}^3$

Para encontrar el espacio nulo buscamos la solución general

$[A_0:\vec{0}]=\left[\begin{matrix}1 & 4 & : & 0\\2 & 8 & : & 0\\-1 & -4& : & 0\end{matrix}\right] \sim \left[\begin{matrix}1 & 4& : & 0\\0 & 0& : & 0\\0 & 0& : & 0\end{matrix}\right]$

* Variables libres:
  * $x_1=t_0$
* Variables delanteras:
  * $1x_0 + 4t_0=0 \ \ \Rightarrow \ \ x_0=-4t_0$
* Solución general
$\left[\begin{matrix}x_0\\x_1\end{matrix}\right]=t_0\left[\begin{matrix}-4\\1\end{matrix}\right]$

Cómo hay una variable libre entonces:
* la nulidad ($\nu$) es 1
* sólo hay un vector en la base del espacio nulo Nu($A_0$), $\left[\begin{matrix}-4\\1\end{matrix}\right]$
* geométricamente el espacio nulo Nu($A_0$) es una recta en $\mathbb{R}^2$



# $A_1$

```python
A1=sp.Matrix([[1,2,3],[4,5,6]])
M=A1[:,:]
M#print(sp.latex(M))
```

```python
M[1,:]=M[1,:]-4*M[0,:]
M#print(sp.latex(M))
```

Para encontrar las columnas que se pueden eliminar buscamos la forma escalón

$A_1=\left[\begin{matrix}1 & 2 & 3\\4 & 5 & 6\end{matrix}\right]\sim \left[\begin{matrix}1 & 2 & 3\\0 & -3 & -6\end{matrix}\right]$

Cómo hay dos columnas con l-pivote (lugar del pivote) entonces:
* el rango ($\rho$) es 2
* hay dos vectores en la base del espacio columna Co($A_1$), $\left[\begin{matrix}1 &2\\4 &5\end{matrix}\right]$
* geométricamente el espacio columna Co($A_1$) es todo $\mathbb{R}^2$

Para encontrar el espacio nulo buscamos la solución general

$[A_1:\vec{0}]=\left[\begin{matrix}1 & 2 & 3 & : & 0\\4 & 5 & 6 & : & 0\end{matrix}\right]\sim \left[\begin{matrix}1 & 2 & 3 & : & 0\\0 & -3 & -6 & : & 0\end{matrix}\right]$

* Variables libres:
  * $x_2=t_0$
* Variables delanteras:
  * $-3x_1  - 6t_0=0 \ \ \Rightarrow \ \ x_1=-2t_0$
  * $1x_0 + 2(-2t_0)  +3t_0=0 \ \ \Rightarrow \ \ x_0=t_0$
* Solución general
$\left[\begin{matrix}x_0\\x_1\\t_2\end{matrix}\right]=t_0\left[\begin{matrix}1\\-2\\1\end{matrix}\right]$

Cómo hay una variable libre entonces:
* la nulidad ($\nu$) es 1
* sólo hay un vector en la base del espacio nulo Nu($A_0$), $\left[\begin{matrix}1\\-2\\1\end{matrix}\right]$
* geométricamente el espacio nulo Nu($A_0$) es una recta en $\mathbb{R}^3$



# $A_2$

```python
A2=sp.Matrix([[1,0,2],[2,1,4],[3,0,6]])
M=A2[:,:]
print(sp.latex(M))
```

```python
M[1,:]=M[1,:]-2*M[0,:]
M[2,:]=M[2,:]-3*M[0,:]
print(sp.latex(M))
```

Para encontrar las columnas que se pueden eliminar buscamos la forma escalón

$A_2=\left[\begin{matrix}1 & 0 & 2\\2 & 1 & 4\\3 & 0 & 6\end{matrix}\right] \sim \left[\begin{matrix}1 & 0 & 2\\0 & 1 & 0\\0 & 0 & 0\end{matrix}\right]$

Cómo hay dos columnas con l-pivote (lugar del pivote) entonces:
* el rango ($\rho$) es 2
* hay dos vectores en la base del espacio columna Co($A_0$), $\left[\begin{matrix}1 & 0 \\2 & 1 \\3 & 0 \end{matrix}\right] $
* geométricamente el espacio columna Co($A_0$) es un plano en $\mathbb{R}^3$

Para encontrar el espacio nulo buscamos la solución general

$[A_2:\vec{0}]=\left[\begin{matrix}1 & 0 & 2 & : & 0\\2 & 1 & 4 & : & 0\\3 & 0 & 6 & : & 0\end{matrix}\right] \sim \left[\begin{matrix}1 & 0 & 2 & : & 0\\0 & 1 & 0 & : & 0\\0 & 0 & 0 & : & 0\end{matrix}\right]$

* Variables libres:
  * $x_2=t_0$
* Variables delanteras:
  * $1x_1  +0t_0=0 \ \ \Rightarrow \ \ x_1=0$
  * $1x_0 + 0  +2t_0=0 \ \ \Rightarrow \ \ x_0=-2t_0$
* Solución general
$\left[\begin{matrix}x_0\\x_1\\t_2\end{matrix}\right]=t_0\left[\begin{matrix}-2\\0\\1\end{matrix}\right]$

Cómo hay una variable libre entonces:
* la nulidad ($\nu$) es 1
* sólo hay un vector en la base del espacio nulo Nu($A_0$), $\left[\begin{matrix}-2\\0\\1\end{matrix}\right]$
* geométricamente el espacio nulo Nu($A_0$) es una recta en $\mathbb{R}^3$
