## 1. **Inversa estándar y generalizadas**

### 1.1 Inversa para una matriz de 2x2

Antes de definir la operación inversa, vamos a resolver un sistema arbitrario de 2 ecuaciones con 2 incógnitas.

$$
\begin{aligned}
&\begin{cases}
ax_1 + bx_2 = & y_1  \\
cx_1 + dx_2 = & y_2
\end{cases}
\end{aligned}
$$

que corresponde a la siguiente matriz extendida en donde se introduce intencionalmente un desfase entre $y_1$ y $y_2$ que será útil para calcular la inversa.

$$
\begin{aligned}
&\begin{bmatrix} a & b & : & y_1 &\\ c & d & : & & y_2 \end{bmatrix} \\
&\vdots \\
\frac{1}{a}R_1 \to R_1
&\begin{bmatrix} 1 & 0 & : & \frac{d}{ad-cb}y_1 &-& \frac{b}{ad-cb}y_2 \\ 0 & 1 & : & -\frac{c}{ad-cb}y_1 &+& \frac{a}{ad-cb}y_2 \end{bmatrix}
\end{aligned}
$$

---

Observe que la solución del sistema de ecuaciones se puede escribir de la siguiente forma:

$$
\begin{aligned}
\begin{bmatrix} x_1 \\ x_2 \end{bmatrix} &= \begin{bmatrix} \frac{d}{ad-cb}y_1 - \frac{b}{ad-cb}y_2 \\ -\frac{c}{ad-cb}y_1 + \frac{a}{ad-cb}y_2 \end{bmatrix} \\
&= \begin{bmatrix} \frac{d}{ad-cb} & \frac{-b}{ad-cb} \\ \frac{-c}{ad-cb} & \frac{a}{ad-cb} \end{bmatrix} \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} \\
&= \frac{1}{ad-cb} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix} \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} \\
&= B \begin{bmatrix} y_1 \\ y_2 \end{bmatrix}, \quad \text{con } B = \frac{1}{ad-cb} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}
\end{aligned}
$$

Si $ ad - bc \neq 0 $, entonces hemos encontrado una matriz $ B $ que permite despejar la ecuación $ A\mathbf{x} = \mathbf{b} $, quedando la ecuación $ \mathbf{x} = B\mathbf{b} $.


---

Observe lo que pasa al multiplicar las dos matrices
 $ A = \begin{bmatrix} a & b \\ c & d \end{bmatrix} $ y $ B = \frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix} $ se obtiene:



$$
AB = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \left(\frac{1}{ad - bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}\right)= \frac{1}{ad - bc}\begin{bmatrix} a & b \\ c & d \end{bmatrix}  \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}= \frac{1}{ad - bc}\begin{bmatrix} ad-bc & -ab+ba \\ cd-dc & -cb+da \end{bmatrix}=  \begin{bmatrix} 1 & 0 \\ 0 & 1 \end{bmatrix}
=I_2
$$

De manera similar se muestra que $BA = I_2$

---
### 1.2 **Definición**
Decimos que si $ A $ es una matriz de tamaño $ n \times n $ y se puede encontrar una matriz $ B $ del mismo tamaño tal que $$ AB = I_n = BA $$ entonces se dice que $ A $ es **invertible** y $ B $ se denomina la **inversa** de $ A $, que se denota $ A^{-1} $.


---

### 1.3 **Cálculo de la inversa utilizando eliminación de Gauss**


El procedimiento para calcular la inversa de la matriz $ A $ se realiza extendiendo la matriz con la matriz identidad, de la siguiente forma:

$$
\begin{aligned}
&\begin{bmatrix} a & b & : & 1 & 0 \\ c & d & : & 0 & 1 \end{bmatrix} \\
R_2 - \frac{c}{a}R_1 \to R_2
&\begin{bmatrix} a & b & : & 1 & 0 \\ 0 & d - \frac{c}{a}b & : & -\frac{c}{a} & 1 \end{bmatrix} \\
\frac{a}{ad-bc}R_2 \to R_2
&\begin{bmatrix} a & b & : & 1 & 0 \\ 0 & 1 & : & -\frac{c}{ad-bc} & \frac{a}{ad-bc} \end{bmatrix} \\
R_1 - bR_2 \to R_1
&\begin{bmatrix} a & 0 & : & \frac{ad}{ad-bc} & \frac{-ab}{ad-bc} \\ 0 & 1 & : & \frac{-c}{ad-bc} & \frac{a}{ad-bc} \end{bmatrix} \\
\frac{1}{a}R_1 \to R_1
&\begin{bmatrix} 1 & 0 & : & \frac{d}{ad-bc} & \frac{-b}{ad-bc} \\ 0 & 1 & : & \frac{-c}{ad-bc} & \frac{a}{ad-bc} \end{bmatrix}
\end{aligned}
$$

En general, para una matriz $A$ de $ n \times n$ se plantea la matriz extendida
$$[A:I_n]$$

Y se busca obtener la forma escalón reducida
$$[I_n:A^{-1}]$$

En este caso $A$ es **invertible** y tiene rango completo. Pero si se obtiene algún renglón de ceros (en las columnas de la mitad izquierda) entonces se dice que la matriz $A$ **no es invertible** y por lo tanto no tiene rango completo.

---

### 1.4 **Propiedades de las matrices invertibles**

Si $ A $ y $ B $ tienen inversa y $ c $ es un escalar distinto de cero, entonces:

- $ A^{-1} $ es única.
- $ (AB)^{-1} = B^{-1} A^{-1} $.
- $ (A^{-1})^{-1} = A $.
- $ (cA)^{-1} = \frac{1}{c} A^{-1} $.
- $ A^T $ es invertible y $ (A^T)^{-1} = (A^{-1})^T $.

### 1.5 Teorema


> La matriz $ A $ de tamaño $ n \times n $ se puede invertir si y solo si el sistema $ A\mathbf{x} = \mathbf{b} $ tiene solución única para cada vector $ \mathbf{b} $.  
> Esta solución única está dada por:  
> $$
> \mathbf{x} = A^{-1} \mathbf{b}
> $$

## 3. **Pseudoinversa**
### 3.1 **Pseudoinversa y Solución de Mínimos Cuadrados sobre dos dimensiones**

Vamos a plantear matricialmente el siguiente sistema de ecuacione

$$
\begin{cases}
(\mathbf{v}_1 \cdot \mathbf{v}_1) k_1 + (\mathbf{v}_1 \cdot \mathbf{v}_2) k_2 = \mathbf{v}_1 \cdot \mathbf{u} \\
(\mathbf{v}_2 \cdot \mathbf{v}_1) k_1 + (\mathbf{v}_2 \cdot \mathbf{v}_2) k_2 = \mathbf{v}_2 \cdot \mathbf{u}
\end{cases}
$$

Para lo cual, para los vectores $\mathbf{v}_1, \mathbf{v}_2, \mathbf{u} \in \mathbb{R}^{n \times 1}$, se necesita definir las siguietes matrices $\mathbf{w}_{2 \times 1}$, $A_{n \times 2}$ :

$$
\mathbf{w} = \begin{bmatrix} k_1 \\ k_2 \end{bmatrix}
$$

La matriz $ A $ está compuesta por los vectores $ \mathbf{v}_1 $ y $ \mathbf{v}_2 $ como columnas:

$$
A =
\begin{bmatrix}
| & |     \\
\mathbf{v}_1 & \mathbf{v}_2  \\
| & |    
\end{bmatrix}
$$

Su transpuesta es:

$$
A^T = \begin{bmatrix}
— \mathbf{v}_1 — \\
— \mathbf{v}_2 — \\
\end{bmatrix}
$$

Multiplicamos $ A^T $ por $ u $:

$$
A^T u = \begin{bmatrix} \mathbf{v}_1^T \\ \mathbf{v}_2^T \end{bmatrix} \mathbf{u} =
\begin{bmatrix} \mathbf{v}_1 \cdot \mathbf{u} \\ \mathbf{v}_2 \cdot \mathbf{u} \end{bmatrix}
$$

La matriz $ A^T A $ es:

$$
A^T A =
\begin{bmatrix} \mathbf{v}_1^T \\ \mathbf{v}_2^T \end{bmatrix}
\begin{bmatrix} \mathbf{v}_1 & \mathbf{v}_2 \end{bmatrix}
=
\begin{bmatrix}
\mathbf{v}_1 \cdot \mathbf{v}_1 & \mathbf{v}_1 \cdot \mathbf{v}_2 \\
\mathbf{v}_2 \cdot \mathbf{v}_1 & \mathbf{v}_2 \cdot \mathbf{v}_2
\end{bmatrix}
$$

De esta forma se escribe matricialmente el anterior sistema de ecuaciones:

$$
\begin{bmatrix}
\mathbf{v}_1 \cdot \mathbf{v}_1 & \mathbf{v}_1 \cdot \mathbf{v}_2 \\
\mathbf{v}_2 \cdot \mathbf{v}_1 & \mathbf{v}_2 \cdot \mathbf{v}_2
\end{bmatrix}
\begin{bmatrix} k_1 \\ k_2 \end{bmatrix}
=
\begin{bmatrix} \mathbf{v}_1 \cdot \mathbf{u} \\ \mathbf{v}_2 \cdot \mathbf{u} \end{bmatrix}
$$

Despejando:

$$
A^T A \mathbf{w} = A^T u
$$

Si $ A^T A $ es invertible, la solución es:

$$
\mathbf{w} = (A^T A)^{-1} A^T u
$$

Este procedimiento es genralizable a más de dos dimensiones.

Esta solución se conoce como **mínimos cuadrados** y ha sido muy usada en regresón y en Machine Learning.

La matriz
$$
A^+ = (A^T A)^{-1} A^T
$$

se conoce como la **pseudoinversa de Moore-Penrose**.

### 3.2 **Propiedades**


1. $ A A^+ A = A $
2. $ A^+ A A^+ = A^+ $
3. $ (A A^+)^T = A A^+ $
4. $ (A^+ A)^T = A^+ A $

- Si $ A $ es una matriz cuadrada e invertible, entonces $ A^+ = A^{-1} $.
- $ A^+ $ proporciona la **solución de mínima norma** a sistemas lineales inconsistentes.


### 3.3 **Cálculo de la Pseudoinversa**

La pseudoinversa de Moore-Penrose se puede calcular de varias maneras, dependiendo de la estructura de $ A $:

#### 3.3.1 Si $ A $ tiene rango completo y es de tamaño $ m \times n $:
   - Si $ A.rows > A.cols $, se usa:

     $$
     A^+ = (A^T A)^{-1} A^T
     $$

   - Si $ A.rows < A.cols $, se usa:

     $$
     A^+ = A^T (A A^T)^{-1}
     $$

#### 3.3.2 Si $ A $ no tiene rango completo:
   En este caso, $ A^T A $ o $ A A^T $ no son invertibles, por lo que se debe utilizar otro método como la descomposición en valores singulares (SVD).


## 4. **La Inversa Generalizada**
La **inversa generalizada** de una matriz $A$ es cualquier matriz $B$ que satisface la ecuación:$$ABA=A$$

A diferencia de la inversa estándar $A^{-1}$, que existe solo cuando A es cuadrada y de rango completo, una inversa generalizada puede existir para cualquier matriz, incluso si no es invertible.

La pseudoinversa de Moore-Penrose $A^{+}$, es un caso especial de inversa generalizada que es única y satisface condiciones adicionales de simetría y mínima norma.
